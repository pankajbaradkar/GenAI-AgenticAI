```python
class Error(Exception):
  pass

class SmallValueError(Error):
  pass

class LargeValueError(Error):
  pass

HiddenNum = 508

while True:

  try:
    inputNum = int(input("Enter a number: "))
    if inputNum < HiddenNum:
      raise SmallValueError

    elif inputNum > HiddenNum:
      raise LargeValueError

    elif inputNum == HiddenNum:
      print("Congratulations! You guessed the correct number.")
    break

  except SmallValueError:
    print("This value is too small, please try again!")

  except LargeValueError:
    print("This value is too large, please try again!")
```
