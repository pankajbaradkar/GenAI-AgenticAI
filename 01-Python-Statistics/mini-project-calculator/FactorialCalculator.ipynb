```python
import os
```

```python
facto = '''def fact(n):
  f=1
  if n>1:
    for i in range(2, n+1):
      f = f * i
  return f'''
```

```python
if not os.path.exists("factorial.py"):
  with open("factorial.py","w") as f:
    f.write(facto)
```
