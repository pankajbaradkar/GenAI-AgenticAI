## 2. f-Strings (Formatted Strings)
f-Strings allow embedding expressions inside string literals using {}.

### Key Benefits
- Cleaner formatting
- Inline expressions
- Supports formatting options


```python
empName = 'Rohit'
empAge = 35
empId = 5454
joinDate = '2026-02-25'
msgToAdmin = f'Employee {empName} with Id {empId} is {empAge} years old and is joining on {joinDate}'
print(msgToAdmin)
```

## 5. Sets
Sets are unordered collections of unique elements.

### Key Characteristics
- Unordered
- No duplicates
- Supports set operations

```python
# Example 1: Create sets and Set Operations
course1 = {'Python', 'Machine Learning', 'Gen AI'}
course2 = {'Machine Learning', "Deep Learning", 'NLP'}
allcourses = course1.union(course2)
print(allcourses)
common = course1.intersection(course2)
print(common)
```

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))
print(a.intersection(b))
```

```python
# Example 3: Adding and Removing
a.add(6)
print(a)
```

```python
a.remove(2)
print(a)
```

## 6. Dictionaries
Dictionaries are unordered collections of key-value pairs.

### Key Characteristics
- Key-value structure
- Mutable
- Keys must be unique and immutable
