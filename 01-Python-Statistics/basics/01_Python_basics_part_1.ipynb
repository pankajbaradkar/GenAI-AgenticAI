## Identifiers/Variable in Python

### Rules for naming an identifier
* A variable name must start with a letter or the underscore character.
* A variable name cannot start with a number.
* A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
* Variable names are case-sensitive (age, Age and AGE are three different variables)

### Logical Operators
Logical Operators are used to combine conditional statements

| Operator | Description | Example | Result |
|----|---|---|--------|
| and  | Returns True if both statements are true | x < 5 and  x < 10	| True |
| or  | Returns True if one of the statements is true | x < 5 or  x < 10 | True |
| not  | Reverse the result, returns False if the result is true and vice versa | not(x < 5 or  x < 10) | False |

---

```python
price=60
quality='good'
if (price<50 and quality=='good'):
  print("I am going to buy")
  print("I am happy")
else:
  print("I am not going to buy")
```

```python
if (price<50 or quality=='good'):
  print("I am going to buy")
  print("I am happy")
else:
  print("I am not going to buy as neither the price is low nor the quality is good")
```

```python
if not(quality=='good'):
  print("I am not going to buy as the quality is not good")
else:
  print('I am going to buy as the quality is good')
```

### Membership Operator
Membership operators are used to test if a sequence is presented in an object

| Operator | Description | Usage | Example | Result |
|----|---|---|----------|--------|
| in  | Returns True if a sequence with the specified value is present in the object| x in y | 3 in [1,2,3] | True |
| not in  | Returns True if a sequence with the specified value is not present in the object| x not in y | 4 not in [1,2,3] | True |

```python
a = [1, 'Rohit', 35]
'Rohit' in a
```

```python
2 in a, 2 not in a
```

```python
35 in a, 35 not in a
```
