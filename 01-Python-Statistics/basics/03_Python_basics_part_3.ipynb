## Classes and Objects in Python (very basic of OOPS)

```python
# a bank account class with deposit, withdraw and check balance methods
class BankAccount:
  '''
  This is a bank account class with deposit, withdraw and check balance methods
  Class: BankAccount
  Methods: deposit, withdraw, checkbalance
  Properties: accno, name, balance
  Constructor: takes accno, name, balance as arguments
  '''
  def __init__(self, accno, name, balance):
    self.accno = accno
    self.name = name
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount
    print("Amount deposited:", amount)
    print("Balance after deposit", self.balance)

  def withdraw(self, amount):  ## Public method
    if self.__checkWithdrawPossible(amount):
      self.balance -= amount
      print("Amount withdrawn:", amount)
      print("Balance after withdrawal", self.balance)
    else:
      print("Insufficient balance")

  def __checkWithdrawPossible(self, amount):  ##Private method
    if self.balance >= amount:
      return True
    else:
      return False

    return self.balance

  def checkbalance(self):
    print("Balance is:", self.balance)
```

```python
rohit = BankAccount('1001', 'Rohit', 10000)
```

```python
print(rohit.checkbalance())
print(rohit.withdraw(3000))
```

```python
rohit.withdraw(15000)
```

```python
rohit.checkbalance()
```

```python
rohit.__checkWithdrawPossible(15000)
```

```python
virat = BankAccount('1002', 'Virat', 20000)
virat.withdraw(15000)
virat.deposit(3000)
```

```python
print(rohit.checkbalance())
print(virat.checkbalance())
```

```python
try:
  a = 10
  b = input("Enter a number:")
  c = a/int(b)
  print(a, b, c)
except Exception as e:
  print("Error while dividing:", e)
```

```python
try:
  a = 10
  b = input("Enter a number:")
  b = int(b)
  c = a/b
  print(a, b, c)
except ZeroDivisionError:
  print("Divide by zero error")

except Exception as e:
  print("Error while dividing:", e)
```

```python
try:
  a = 10
  b = input("Enter a number:")
  b = int(b)
  c = a/b
  print(a, b, c)
except ZeroDivisionError:
  print("Divide by zero error")
except ValueError:
  print("Invalid input")
except Exception as e:
  print("Error while dividing:", e)
```

```python
class Error(Exception):
  pass

class ValueSmallError(Error):
  pass

class ValueLargeError(Error):
  pass

number = 108
while True:
  try:
    in_num = int(input("Enter a number: "))
    if in_num < number:
      raise ValueSmallError
    elif in_num > number:
      raise ValueLargeError
    elif in_num == number:
      print("You guessed it right")
      break
  except ValueSmallError:
    print("This value is too small, try again")
    print()
  except ValueLargeError:
    print("This value is too large, try again")
    print()

print("Congratualtion...you guessed it correct")
```
